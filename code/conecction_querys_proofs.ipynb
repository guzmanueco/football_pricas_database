```python
# Importing requests and BeautifulSoup for scrapping data from transfer market

import requests
from bs4 import BeautifulSoup
```

```python
# Creating header for the site to think we are a browser and not scrappers

headers = {'User-Agent': 
           'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}
```

```python
# Setting the url address we want to scrap

url = "https://www.transfermarkt.co.uk/transfers/transferrekorde/statistik/top/plus/0/galerie/0?saison_id=2019"
```

```python
# Doing the request
urlrequest = requests.get(url, headers=headers)
```

```python
# By calling the variable urlrequest we get a response 200, meaning we have obtained a poisvite answer to our
# request. However we can not see the data yet. For that, we would have to parse it
urlrequest
```

```python
# We are parsing the data into html code to be able to scrap through it

urlSoup = BeautifulSoup(urlrequest.content, 'html.parser')
```
